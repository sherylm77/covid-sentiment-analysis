```python
import csv
import time
#!pip3 install google-cloud-translate==2.0.1
from google.cloud import translate_v2 as translate

translate_client = translate.Client()

tsv_file = "Tweet Annotations - May.tsv"
data = open(tsv_file, "r", encoding='utf-8')
read_tsv = csv.reader(data, delimiter="\t")

chars = 0
row_count = 1

with open('Tweet Annotations - MayTranslated.tsv', 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    
    for row in read_tsv:
        if row_count <= 4325:
            row_count += 1
            continue
        #try:
        if row[3] != "en":
            chars += len(row[2])
            text_to_translate = row[2]
            language = row[3]
            print(language, text_to_translate)
            result = translate_client.translate(text_to_translate, target_language='en')
            tsv_writer.writerow([row[0], row[1], result['translatedText'], row[3], row[4], row[5]])
        else:
            tsv_writer.writerow(row)
        #except:
            #print("error CHARS COUNT", chars)
            #print("error row count", row_count)

        print("CHARS COUNT", chars)
        print("row count", row_count)
        row_count += 1
```
